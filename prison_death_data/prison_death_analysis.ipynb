```python
us_death_rates_df = pd.read_csv("US_Death_Rates.csv")
```

```python
us_death_rates_df.info()
```

```python
##Reading in death rates for states general populations
us_death_rates_df
```

```python
##Fixing spelling
us_death_rates_df = us_death_rates_df.rename(columns={"State homocide": "State homicide"})
us_death_rates_df
```

```python
rate_state_dc_df.info()
```

```python
##Combining death rates for prison pops and general pops. "State" is general population.
combined_df = pd.concat([rate_state_dc_df, us_death_rates_df], axis=1)           
```

```python
combined_df
```

```python
##Comparing prison and state suicide rates
suicide_combined_df = combined_df.filter(["Jurisdiction", "Suicide", "State suicide"], axis=1)
suicide_combined_df
```

```python
##Adding difference column
suicide_combined_df["difference"] = suicide_combined_df["Suicide"] - suicide_combined_df["State suicide"]
suicide_combined_df
```

```python
##Most states with big disparities between prison suicide rates and general public rates are smaller 
## like Utah, Rhode Island, Delaware, etc. But New York and Mass have high prison suicide rates.
suicide_combined_df.sort_values(by=["difference"], ascending = False)
```

```python
suicide_combined_df["Suicide"].mean()
```

```python
suicide_combined_df["State suicide"].mean()
```

```python
suicide_combined_df.loc[31]
##NY's discrepancy is due more to a low suicide rate for the general public than a high suicide rate for prisoners
##The state is only a couple notches above the prison rate, but half the general public rate.
```

```python
liver_combined_df = combined_df.filter(["Jurisdiction", "Liver disease", "State liver disease"], axis=1)
liver_combined_df
```

```python
liver_combined_df["difference"] = liver_combined_df["Liver disease"] - liver_combined_df["State liver disease"]
liver_combined_df
```

```python
liver_combined_df["Liver disease"].mean()
```

```python
liver_combined_df["State liver disease"].mean()
```

```python
homicide_combined_df = combined_df.filter(["Jurisdiction", "Homicide", "State homicide"], axis=1)
homicide_combined_df
```

```python
homicide_combined_df["difference"] = homicide_combined_df["Homicide"] - homicide_combined_df["State homicide"]
homicide_combined_df
```

```python
##Comparing homocide rates in state prisons vs overall state homicide rates.
homicide_combined_df.sort_values(by=["difference"], ascending = False)
```
